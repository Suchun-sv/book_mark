# 消除后面的...变成...

```python
for  number, line in enumerate(context):
    if len(line) > 5:
        new_line = line[5: -1]
        if '.' in new_line:
            new_line = new_line.replace('.', '…')
#             print(new_line)
            context[number] = context[number][0:5] + new_line
    else:
        pass
```

# 消除 …

```python
for  number, line in enumerate(context):
    if '…' in line:
        while line.count('…') != 1:
            line = line.replace('…', '',  1)
        context[number] = line.replace('…', '\t')
```

```python
context
```

# tab

```python
for num, line in enumerate(context):
    N = line.count('.')
    context[num] = '\t'*N + context[num]
```

# 页码确定

```python
for num, line in enumerate(context):
    try:
        if line[-1] != '\n':
            context[num] = context[num]+'\n'
    except:
        continue
```

```python
for num, line in enumerate(context):
    line = line[0: -1]
#     print(line.split('\t'))
    div = line.split('\t')
    if div[-1] == '':
        div = context[num-1][0:-1].split('\t')[-1]
    try:
        page = int(div[-1])
    except:
        continue
#     print(page)
    context[num] = div[0]
    for i in range(1, len(div)):
        context[num] += '\t'+ div[i]
    context[num]+='\n'
```

# 文件写入

```python
with open('new_bookmark.txt', 'w') as f:
    for i in context:
        f.write(i)
```
